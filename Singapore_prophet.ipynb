```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
data = pd.read_csv('drive/MyDrive/dry/modeling/data/Singapore/total.csv')
keys = data.columns
data[keys[0]] = pd.to_datetime(data[keys[0]].astype(str) + '-' + data[keys[1]].astype(str), format = '%Y-%m')
data = data.drop(keys[1], axis = 1)
keys = data.columns
# data
```

```python
fig, ax = plt.subplots()
color = ["blue","pink","red","black"]
ax.set_xlabel('date')
ax.set_ylabel('ratio(%)')
ax.grid(axis = 'y')

for i in range(1, 5):
    ax.plot(data[keys[0]], data[keys[i]], color = color[i - 1])
plt.show()
```

```python
denv = [data.iloc[:, [0, i]].rename(columns = {'year' : 'ds', keys[i] : 'y'}) for i in range(1, 5)]
keys_denv = denv[0].columns
```

```python
train_data = [train_test_split(denv[i], shuffle = False, test_size = 0.1)[0] for i in range(4)]
test_data = [train_test_split(denv[i], shuffle = False, test_size = 0.1)[1] for i in range(4)]
```

```python
def model(num):
    m = Prophet()
    m.fit(train_data[num])
    forecast = m.predict(pd.DataFrame(test_data[num][keys_denv[0]]))
    mae = mean_absolute_error(forecast['yhat'], test_data[num][keys_denv[1]])
    # print(f'MAE = {mae}')

    fig = plt.figure(figsize = (12, 8))
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(denv[num][keys_denv[0]], denv[num][keys_denv[1]], color = 'y')
    m.plot(forecast, ax = ax)
    m.plot_components(forecast)
    ax.set_xlabel('date')
    ax.set_ylabel(f'DENV-{num + 1} ratio(%)')
    plt.show()

    return forecast
```

```python
forecasts = []
for i in range(4):
    forecasts.append(model(i))
```
